```python
import pandas as pd
import numpy as np
df = pd.read_csv("placement_data.csv")
students = len(df["Name"])
print("Total number of students: ",students)
avg_cgpa = np.average(df["CGPA"])
print("Average CGPA: ",avg_cgpa)
highest_cgpa = df.loc[df["CGPA"].idxmax()]
print("Highest CGPA Student: ",highest_cgpa["Name"])
count = len(df[df["Placed"] == "Yes"])
print("Total placed students: ",count)
avg_package = np.average(df["Package_LPA"])
print("Average CGPA: ",avg_package)
highest_package = df.loc[df["Package_LPA"].idxmax()]
print("Highest Package Student: ",highest_package["Name"])
highest_package_student = np.max(df["Package_LPA"])
print(f"Highest Package: {highest_package_student} LPA")
placed = df[df["Placed"] == "Yes"]
print(placed["Department"].value_counts())
```
