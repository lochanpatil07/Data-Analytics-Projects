```python
import pandas as pd
import numpy as np
df = pd.read_csv("employee_data.csv")

print("Total Employees: ",len(df["Name"]))
print("Average Salary: ",np.average(df["Salary"]))
highest = df.loc[df["Salary"].idxmax()]
print("Highest Salary Employee: ",highest["Name"])
lowest = df.loc[df["Salary"].idxmin()]
print("Lowest Salary Employee: ",lowest["Name"])

#Filtering
print("\nEmployees with experience > 3 years: ")
exp = df[df["Experience"] >= 3]
print(exp[["Name", "Salary"]])

print("\nEmployees with good performance: ")
good = df[df["Performance"] == "Good"]
print(good[["Name", "Performance"]])

print("\nEmployees earning salary b/w 40000 and 70000: ")
salary_range = df[df["Salary"] >= 40000 and df["Salary"] <= 70000]
print(salary_range[["Name", "Salary"]])

print("\nMechanical department employees: ")
mechanical = df[df["Department"] == "Mechanical"]
print(mechanical[["Name", "Department"]])

#Sorting
df.sort_values(by="Salary", ascending=False)
df.sort_values(by="Experience", ascending=False)

#Grouping
print(df.groupby("Department")["Salary"].mean())
print(df["Department"].value_counts())
#or print(df.groupby("Department").size())
```
